```python
import numpy as np
import pandas as pd
# Initialize empty lists to store t and y values
t_data = []
y_data = []

# Read the file line by line
file_path = r"C:\Users\ashut\Downloads\data-44.txt"  # Replace with your actual file path
with open(file_path, 'r') as file:
    for line in file:
        # Split the line by any whitespace (including multiple spaces)
        values = line.split()
        
        # Convert the values to floats and append them to the lists
        t_data.append(float(values[0]))
        y_data.append(float(values[1]))

# Convert lists to numpy arrays
t_data = np.array(t_data)
y_data = np.array(y_data)

# Check the extracted data
print("t_data:", t_data)
print("y_data:", y_data)

```

```python

from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define each model as a function

# Model 1: y(t) = α0 + α1 * e^(β1 * t) + α2 * e^(β2 * t) + ε(t)
def model_1(t, a0, a1, b1, a2, b2):
    return a0 + a1 * np.exp(b1 * t) + a2 * np.exp(b2 * t)

# Model 2: y(t) = (α0 + α1 * t) / (β0 + β1 * t) + ε(t)
def model_2(t, a0, a1, b0, b1):
    return (a0 + a1 * t) / (b0 + b1 * t)

# Model 3: y(t) = β0 + β1 * t + β2 * t^2 + β3 * t^3 + β4 * t^4 + ε(t)
def model_3(t, b0, b1, b2, b3, b4):
    return b0 + b1 * t + b2 * t**2 + b3 * t**3 + b4 * t**4



# Fit each model to the data
params_1, _ = curve_fit(model_1, t_data, y_data, maxfev=10000)
params_2, _ = curve_fit(model_2, t_data, y_data, maxfev=10000)
params_3, _ = curve_fit(model_3, t_data, y_data, maxfev=10000)

# Predict using the fitted models
y_pred_1 = model_1(t_data, *params_1)
y_pred_2 = model_2(t_data, *params_2)
y_pred_3 = model_3(t_data, *params_3)

# Plot the results
plt.figure(figsize=(12, 8))
plt.scatter(t_data, y_data, color='black', label='Data', s=10)
plt.plot(t_data, y_pred_1, label='Model 1 Fit', color='blue')
plt.plot(t_data, y_pred_2, label='Model 2 Fit', color='green')
plt.plot(t_data, y_pred_3, label='Model 3 Fit', color='red')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.title('Nonlinear Regression Models')
plt.show()
```

```python

from scipy.stats import shapiro

# Define models
def model_1(t, a0, a1, a2, b1, b2):
    return a0 + a1 * np.exp(b1 * t) + a2 * np.exp(b2 * t)

def model_2(t, a0, a1, b0, b1):
    return (a0 + a1 * t) / (b0 + b1 * t)

def model_3(t, b0, b1, b2, b3, b4):
    return b0 + b1 * t + b2 * t**2 + b3 * t**3 + b4 * t**4

# Set initial guesses for each model
initial_guess_1 = [1, 1, 1, 1, 1]  # Modify these values as needed
initial_guess_2 = [1, 1, 1, 1]
initial_guess_3 = [1, 1, 1, 1, 1]

# Fit each model and calculate residuals
params_1, _ = curve_fit(model_1, t_data, y_data, p0=initial_guess_1, maxfev=5000, bounds=(0, np.inf))
params_2, _ = curve_fit(model_2, t_data, y_data, p0=initial_guess_2, maxfev=5000, bounds=(0, np.inf))
params_3, _ = curve_fit(model_3, t_data, y_data, p0=initial_guess_3, maxfev=5000, bounds=(0, np.inf))

y_pred_1 = model_1(t_data, *params_1)
y_pred_2 = model_2(t_data, *params_2)
y_pred_3 = model_3(t_data, *params_3)

residuals_1 = y_data - y_pred_1
residuals_2 = y_data - y_pred_2
residuals_3 = y_data - y_pred_3

# Calculate Mean Squared Error (MSE) for each model to find the best model
mse_1 = np.mean(residuals_1**2)
mse_2 = np.mean(residuals_2**2)
mse_3 = np.mean(residuals_3**2)

print("MSE Model 1:", mse_1)
print("MSE Model 2:", mse_2)
print("MSE Model 3:", mse_3)

# Choose the best model based on the smallest MSE
best_model = min((mse_1, 'Model 1'), (mse_2, 'Model 2'), (mse_3, 'Model 3'))
print("Best model:", best_model)

# Confidence intervals can be derived from the covariance matrix, omitted here for simplicity

# Plot residuals for the best model
plt.figure()
plt.title('Residuals of model 1')
plt.plot(t_data, residuals_1, 'o')
plt.xlabel('t')
plt.ylabel('Residuals')
plt.show()
print("\n")
plt.title('Residuals of model 2')
plt.plot(t_data, residuals_2, 'o')
plt.xlabel('t')
plt.ylabel('Residuals')
plt.show()
print("\n")
plt.title('Residuals of model 3')
plt.plot(t_data, residuals_3, 'o')
plt.xlabel('t')
plt.ylabel('Residuals')
plt.show()


# Test for normality of residuals (Shapiro-Wilk test)
residuals_best = residuals_1 if best_model[1] == 'Model 1' else (residuals_2 if best_model[1] == 'Model 2' else residuals_3)
shapiro_test = shapiro(residuals_best)
print("Shapiro-Wilk test for normality of residuals:", shapiro_test)

# Plot the observed data points and the fitted curve of the best model
plt.figure()
plt.scatter(t_data, y_data, label='Observed data', color='black')
if best_model[1] == 'Model 1':
    plt.plot(t_data, y_pred_1, label='Fitted curve (Model 1)', color='blue')
elif best_model[1] == 'Model 2':
    plt.plot(t_data, y_pred_2, label='Fitted curve (Model 2)', color='blue')
else:
    plt.plot(t_data, y_pred_3, label='Fitted curve (Model 3)', color='blue')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
from scipy.stats import norm

# Define your model here (for example, model_1 as shown in earlier steps)

# Fit the model and obtain parameters
params, covariance = curve_fit(model_1, t_data, y_data, p0=initial_guess_1, maxfev=5000, bounds=(0, np.inf))
sigma_squared_estimate = np.sum((y_data - model_1(t_data, *params))**2) / (len(y_data) - len(params))

# Calculate the Fisher Information Matrix approximation
fisher_info_matrix = covariance / sigma_squared_estimate

# Standard errors of the parameters (square root of diagonal of covariance matrix)
standard_errors = np.sqrt(np.diag(fisher_info_matrix))

# Confidence intervals (95%)
alpha = 0.05  # for 95% confidence interval
z_value = norm.ppf(1 - alpha / 2)  # approximately 1.96 for 95% confidence
confidence_intervals = [(param - z_value * se, param + z_value * se) for param, se in zip(params, standard_errors)]

# Display results
for i, (param, ci) in enumerate(zip(params, confidence_intervals)):
    print(f"Parameter {i}: Estimate = {param:.4f}, 95% Confidence Interval = [{ci[0]:.4f}, {ci[1]:.4f}]")
```
